```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Load in DataFrame
stroke_df = pd.read_csv("Resources/adult.csv")
stroke_df.head()
```

```python
#remove "ghotst" column
stroke_df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Export Selected Columns to CSV Files
independent_columns = ['age', 'gender', 'ever_married']
stroke_df.to_csv("Resources/independent.csv", index=False, columns=independent_columns)
```

```python
dependent_columns = ['stroke','hypertension','heart_disease','bmi','avg_glucose_level']
stroke_df.to_csv("Resources/dependent.csv", index=False, columns=dependent_columns)
```

```python
employment_columns = ['age','gender','stroke','bmi','avg_glucose_level','ever_married','govt_worker','never_worked','private_worker','self_employed','minor_unemployed']
stroke_df.to_csv("Resources/employment.csv", index=False, columns=employment_columns)
```

```python
cardio_columns = ['age','gender','stroke','bmi','avg_glucose_level','ever_married','hypertension','heart_disease']
stroke_df.to_csv("Resources/cardiovascular.csv", index=False, columns=cardio_columns)
```

```python
housing_columns = ['age','gender','stroke','bmi','avg_glucose_level','ever_married','rural_residence','urban_residence']
stroke_df.to_csv("Resources/housing.csv", index=False, columns=housing_columns)
```

```python
smoking_columns = ['age','gender','stroke','bmi','avg_glucose_level','ever_married','smoking_unknown','former_smoker','never_smoked','current_smoker']
stroke_df.to_csv("Resources/smoking.csv", index=False, columns=smoking_columns)
```
