```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Load in DataFrame
stroke_df = pd.read_csv("Resources/adult.csv")
stroke_df.head()
```

```python
# Export Selected Columns to CSV Files
individual_columns = ['id','age','gender','ever_married']
stroke_df.to_csv("Resources/individual.csv", index=False, columns=individual_columns)
```

```python
employment_columns = ['id','govt_worker','never_worked','private_worker','self_employed']
stroke_df.to_csv("Resources/employment.csv", index=False, columns=employment_columns)
```

```python
cardio_columns = ['id','stroke','bmi','avg_glucose_level','hypertension','heart_disease']
stroke_df.to_csv("Resources/cardiovascular.csv", index=False, columns=cardio_columns)
```

```python
housing_columns = ['id','rural_residence','urban_residence']
stroke_df.to_csv("Resources/housing.csv", index=False, columns=housing_columns)
```

```python
smoking_columns = ['id','smoking_unknown','former_smoker','never_smoked','current_smoker']
stroke_df.to_csv("Resources/smoking.csv", index=False, columns=smoking_columns)
```
