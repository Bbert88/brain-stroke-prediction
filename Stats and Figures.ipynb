```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Load in DataFrame
stroke_df = pd.read_csv("Resources/adult.csv")
stroke_df.head()
```

```python
stroke_df.info()
```

```python
stroke_df.describe()
```

```python
# Create the plot.
x_axis = stroke_df['bmi']
y_axis = stroke_df['age']

plt.bar(x_axis, y_axis, color="blue", linewidth=2, label='Age vrs BMI')
# Create labels for the x and y axes.
plt.xlabel("BMI")
plt.ylabel("Age")
# Create a title.
plt.title("Age vrs BMI")
# Add a grid.
plt.grid()
```

```python
plt.barh(x_axis, y_axis)
# Create labels for the x and y axes.
plt.xlabel("BMI")
plt.ylabel("Age")
# Create a title.
plt.title("Age vrs BMI")
# Add a grid.
plt.grid()
```
